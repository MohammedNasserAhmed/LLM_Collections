ðŸ”³ ***Import Necessary Libs:***

Import the essential libraries like pathlib, textwrap, genai (for GenAI functionalities), and display libraries from IPython to present results. Additionally, getpass and os are used for secure API token input and environment variable management.

```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown

from getpass import getpass
import os
```

ðŸ”³ Authentication (Secure Tokens):

* The code utilizes getpass to securely prompt you for your Google API token. This token is crucial for accessing GenAI services.

* It then sets the GOOGLE_API_TOKEN environment variable using os.environ and retrieves the stored value using os.getenv.

* Finally, it configures GenAI with the obtained API key using genai.configure.

```python
api_token = getpass(prompt='Enter your Google API token: ')
os.environ['GOOGLE_API_TOKEN'] = api_token
GOOGLE_API_TOKEN = os.getenv('GOOGLE_API_TOKEN')
genai.configure(api_key=GOOGLE_API_TOKEN)
```

ðŸ”³ ***List Available Models:***

* The code iterates through GenAI's available models (genai.list_models()) and filters for models that support the generateContent method.

* The names of suitable models are then printed to the console.

```python
for m in genai.list_models():
  if 'generateContent' in m.supported_generation_methods:
    print(m.name)
```

ðŸ”³ ***Create draft prompts:***

Set some prompts as you wish.

```python
prompt1 = "There exists a unique increasing geometric sequence of five 2-digit positive integers. What is their sum?."
prompt2 = "Let  k,l>0 be parameters. The parabola  y=kx2âˆ’2kx+l intersects the line  y=4 at two points  A and B . These points are distance 6 apart. What is the sum of the squares of the distances from  A and  B to the origin?"
```

ðŸ”³ ***Select a Model:***

* The code explicitly selects the gemini-pro model from GenAI's offerings using genai.GenerativeModel.

* This model is a powerful LLM capable of generating various creative text formats.

```python
model = genai.GenerativeModel('gemini-pro')
```

ðŸ”³ ***Generate Text with Explanation (%%time Cell)***:

* This code cell is marked with %%time, which indicates that the execution time will be measured and displayed.

* The core functionality resides within model.generate_content. This line prompts the gemini-pro model to generate text based on the provided prompt (f"{prompt1}").


```python
%%time
response = model.generate_content(f"{prompt1}")
```

ðŸ”³ ***Display Generated Text:***

The response.text attribute holds the generated text by the model. *Below code retrieves and displays it.*

```python
response.text
```

ðŸ”³ ***Formatting for Readability (Optional):***

* The provided function to_markdown is for optional formatting. It replaces bullet points with asterisks and indents the text for better readability.

* The formatted text is then displayed using Markdown from IPython.display.

```python
def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
to_markdown(response.text)
```
